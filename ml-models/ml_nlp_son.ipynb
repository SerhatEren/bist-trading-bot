```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Bidirectional, LSTM, Dense, Dropout, Layer
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt


gpus = tf.config.list_physical_devices('GPU')
print("Kullanılabilir GPU sayısı:", len(gpus))


df = pd.read_csv("THYAO Geçmiş Verileri (2).csv", delimiter=",")


#  en yeni tarih varsa verileri ters çevir
df = df.iloc[::-1].reset_index(drop=True)

# "Tarih","Şimdi","Açılış","Yüksek","Düşük","Hac.","Fark %"
def parse_number(x):
    if isinstance(x, str):
        x = x.strip('"')
        x = x.replace(',', '.')
        if 'M' in x:
            x = x.replace('M', '')
            return float(x) * 1e6
        elif 'B' in x:
            x = x.replace('B', '')
            return float(x) * 1e9
    return float(x)

for col in ["Şimdi", "Açılış", "Yüksek", "Düşük", "Hac."]:
    df[col] = df[col].apply(parse_number)

df["Fark %"] = df["Fark %"].apply(lambda x: float(x.replace('%', '').replace(',','.')))

# Tarih sütununu formatla ve artan sıraya göre sırala
df["Tarih"] = pd.to_datetime(df["Tarih"], format='%d.%m.%Y')
df = df.sort_values("Tarih", ascending=True).reset_index(drop=True)


# teknik göstergeler

def compute_RSI(series, period=14):
    delta = series.diff()
    gain = delta.where(delta > 0, 0)
    loss = -delta.where(delta < 0, 0)
    avg_gain = gain.rolling(window=period, min_periods=period).mean()
    avg_loss = loss.rolling(window=period, min_periods=period).mean()
    RS = avg_gain / avg_loss
    RSI = 100 - (100 / (1 + RS))
    return RSI

def compute_MACD(series, fast=12, slow=26, signal=9):
    ema_fast = series.ewm(span=fast, adjust=False).mean()
    ema_slow = series.ewm(span=slow, adjust=False).mean()
    macd = ema_fast - ema_slow
    macd_signal = macd.ewm(span=signal, adjust=False).mean()
    return macd, macd_signal


df['RSI'] = compute_RSI(df["Şimdi"])
df['MACD'], df['MACD_signal'] = compute_MACD(df["Şimdi"])
df['SMA10'] = df["Şimdi"].rolling(window=10).mean()
df = df.dropna().reset_index(drop=True)


feature_cols = ["Şimdi", "Açılış", "Yüksek", "Düşük", "Hac.", "RSI", "MACD", "MACD_signal", "SMA10"]
features = df[feature_cols].values
target = df["Fark %"].shift(-1).dropna().values
features = features[:-1]


scaler = MinMaxScaler()
features_scaled = scaler.fit_transform(features)


sequence_length = 20  #  sliding window

def create_sequences(features, target, seq_length):
    X, y = [], []
    for i in range(len(features) - seq_length):
        X.append(features[i:i+seq_length])
        y.append(target[i+seq_length])
    return np.array(X), np.array(y)

X, y = create_sequences(features_scaled, target, sequence_length)


split = int(0.9 * len(X))
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]



class Attention(Layer):
    def __init__(self, **kwargs):
        super(Attention, self).__init__(**kwargs)

    def build(self, input_shape):
        # input_shape: (batch_size, time_steps, features)
        self.W = self.add_weight(name="att_weight", shape=(input_shape[-1], input_shape[-1]),
                                 initializer="glorot_uniform", trainable=True)
        self.b = self.add_weight(name="att_bias", shape=(input_shape[-1],),
                                 initializer="zeros", trainable=True)
        super(Attention, self).build(input_shape)

    def call(self, x):
        # x: (batch_size, time_steps, features)
        e = tf.keras.backend.tanh(tf.keras.backend.dot(x, self.W) + self.b)
        a = tf.keras.backend.softmax(e, axis=1)
        output = tf.keras.backend.sum(x * a, axis=1)
        return output

# boyutlar
time_steps = X_train.shape[1]    # sequence_length
num_features = X_train.shape[2]


input_seq = Input(shape=(time_steps, num_features))
x = Bidirectional(LSTM(128, return_sequences=True))(input_seq)
x = Dropout(0.3)(x)
x = Bidirectional(LSTM(128, return_sequences=True))(x)
x = Dropout(0.3)(x)
x = Bidirectional(LSTM(64, return_sequences=True))(x)
x = Dropout(0.3)(x)
att_out = Attention()(x)
dense1 = Dense(64, activation='relu')(att_out)
dense2 = Dense(32, activation='relu')(dense1)
output = Dense(1, activation='linear')(dense2)

model = Model(inputs=input_seq, outputs=output)
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001), loss='mse')
print(model.summary())


callbacks = [
    EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True),
    ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=5, min_lr=1e-6)
]


history = model.fit(X_train, y_train, epochs=100, batch_size=32, validation_split=0.1, callbacks=callbacks)
model.save("3xB-LSTM_rnn_model.h5")
print("model kaydedildi: 3xB-LSTM_rnn_model.h5")


test_loss = model.evaluate(X_test, y_test)
print("Test kaybı (loss):", test_loss)




predictions = model.predict(X_test)

# Test verisinin orijinal DataFrame karşılığını, sequence oluşturma sonucu elde edilen örnek sayısı kadar al
df_test = df.iloc[split + sequence_length : split + sequence_length + len(X_test)].reset_index(drop=True)
actual_prices = df_test["Şimdi"].values
dates = df_test["Tarih"].values


# önceki günün gerçek fiyatına göre yüzdelik değişimi hesapla
predicted_prices = []
for i, pred in enumerate(predictions):
    if i == 0:
        base_price = df.iloc[split + sequence_length - 1]["Şimdi"]
    else:
        base_price = df_test["Şimdi"].iloc[i - 1]
    predicted_price = base_price * (1 + pred[0] / 100)
    predicted_prices.append(predicted_price)

plt.figure(figsize=(14, 7))
plt.plot(dates, actual_prices, label="Gerçek Fiyat", color="blue")
plt.plot(dates, predicted_prices, label="Tahmin Edilen Fiyat", color="orange", linestyle="--")
plt.xlabel("Tarih")
plt.ylabel("Fiyat")
plt.legend()
plt.show()





actual_returns = df.iloc[split + sequence_length + 1 : split + sequence_length + 1 + len(X_test)]["Fark %"].values
print("Test örnek sayısı:", len(X_test), " / Gerçek yüzde değişim örnek sayısı:", len(actual_returns))

initial_capital = 10000.0
capital = initial_capital
transactions = []


for i in range(len(X_test)):
    # model tahmini >= 0 ise long, < 0 ise short
    signal = 1 if predictions[i][0] >= 0 else -1

    trade_return = (actual_returns[i] / 100.0) if signal == 1 else -(actual_returns[i] / 100.0)
    capital_before = capital
    capital = capital * (1 + trade_return)
    trade_date = df.iloc[split + sequence_length + 1 + i]["Tarih"]
    transactions.append((trade_date, signal, actual_returns[i], capital_before, capital))

print("İlk sermaye:", initial_capital)
print("Son sermaye:", capital)
print("Toplam işlem sayısı:", len(transactions))

# işlem geçmişi
transactions_df = pd.DataFrame(transactions, columns=["Tarih", "Signal", "Actual_Return(%)", "Capital_Before", "Capital_After"])
print(transactions_df.head())


dates_trading = transactions_df["Tarih"]
capitals = transactions_df["Capital_After"]
plt.figure(figsize=(14, 7))
plt.plot(dates_trading, capitals, marker="o", color="purple")
plt.xlabel("Tarih")
plt.ylabel("Total")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
import feedparser
import datetime
import torch
from transformers import AutoTokenizer, AutoModelForSequenceClassification

# Fine-tuned 3 sınıflı modelin yükleneceği dizin
model_path = "/content/drive/MyDrive/fine_tuned_3class_model"
tokenizer = AutoTokenizer.from_pretrained(model_path)
model = AutoModelForSequenceClassification.from_pretrained(model_path)

# Model çıktı etiketlerini düzenleyelim:
# (0 = Negatif, 1 = Nötr, 2 = Pozitif)
model.config.id2label = {0: "Negatif", 1: "Nötr", 2: "Pozitif"}


# RSS besleme URL'si (Hürriyet Ekonomi)
rss_url = "https://www.hurriyet.com.tr/rss/ekonomi/"
feed = feedparser.parse(rss_url)

def analyze_news(feed, tokenizer, model, max_articles=5):
    """
    RSS beslemesinden gelen haberleri alır, başlık ve özet bilgilerini birleştirir,
    her haber için duygu analizi yapar.
    """
    news_results = []
    for entry in feed.entries[:max_articles]:
        # Yayın tarihini alıp datetime formatına çeviriyoruz
        news_date = entry.get("published", "")
        try:
            news_date = datetime.datetime(*entry.published_parsed[:6])
        except Exception:
            news_date = None

        title = entry.get("title", "")
        summary = entry.get("summary", "")
        # Haber metnini başlık ve özet olarak birleştiriyoruz.
        text = title + ". " + summary

        # Metni tokenize edip modele uygun hale getiriyoruz.
        inputs = tokenizer(text, return_tensors="pt")
        outputs = model(**inputs)
        # Çıktıdaki logits değerlerinden softmax ile olasılıkları hesaplıyoruz.
        predictions = torch.nn.functional.softmax(outputs.logits, dim=-1)
        pred_idx = predictions.argmax(dim=-1).item()
        # Güncellenen id2label sayesinde etiket "Negatif", "Nötr", "Pozitif" olarak görünecek.
        label = model.config.id2label[pred_idx] if hasattr(model.config, "id2label") else str(pred_idx)
        score = predictions[0, pred_idx].item()

        sentiment = {"label": label, "score": score}
        news_results.append({
            "date": news_date,
            "title": title,
            "summary": summary,
            "sentiment": sentiment
        })
    return news_results

# Son 5 haberi analiz ediyoruz.
news_results = analyze_news(feed, tokenizer, model, max_articles=5)

# Sonuçları yazdırıyoruz.
for news in news_results:
    print("Tarih:", news["date"])
    print("Başlık:", news["title"])
    print("Özet:", news["summary"])
    print("Duygu:", news["sentiment"])
    print("-" * 80)
```
